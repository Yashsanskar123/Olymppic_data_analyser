```python
def fetch_medal_tally(df,year='Overall', country='Overall'):
    medal_df = df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])    
    flag = 0
    temp_df = medal_df.copy()  # Initialize temp_df to a copy of medal_df

    if year != 'Overall' and country != 'Overall':
        temp_df = medal_df[(medal_df['Year'] == int(year)) & (medal_df['region'] == country)]
    elif year != 'Overall':
        temp_df = medal_df[medal_df['Year'] == int(year)]
    elif country != 'Overall':
        flag = 1
        temp_df = medal_df[medal_df['region'] == country]
        
    if flag == 1:
        x = temp_df.groupby('Year').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Year').reset_index()
    else:
        x = temp_df.groupby('region').sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending=False).reset_index()

    x['total'] = x['Gold'] + x['Silver'] + x['Bronze']
    
    print(x)
```

```python
def most_successful(df, sport):
    
    temp_df = df.dropna(subset=['Medal'])
    
    
    if sport != 'Overall':
        temp_df = temp_df[temp_df['Sport'] == sport]
        
    
    top_athletes = temp_df['Name'].value_counts().reset_index().head(15)
    top_athletes.columns = ['Name', 'MedalCount']  
    

    merged_df = top_athletes.merge(df, left_on='Name', right_on='Name', how='left')
    
    
    result_df = merged_df[[ 'Name', 'Sport', 'region']]
    result_df = result_df.drop_duplicates()
    
    return result_df


most_successful(df, 'Overall')
```

```python
def most_successful(df,country):
    temp_df = df.dropna(subset=['Medal'])
    temp_df =  temp_df[temp_df['region'] == country]
    
    top_athletes = temp_df['Name'].value_counts().reset_index().head(15)
    top_athletes.columns = ['Name', 'MedalCount']
    merged_df = top_athletes.merge(df, left_on='Name', right_on='Name', how='left')
    result_df = merged_df[[ 'Name', 'Sport','MedalCount']]
    result_df = result_df.drop_duplicates()
    
    return result_df
```

```python
athlete_df = df.drop_duplicates(subset=['Name', 'region'])
gold_medalists = athlete_df[athlete_df['Medal'] == 'Gold']
silver_medalists = athlete_df[athlete_df['Medal'] == 'Silver']
bronze_medalists = athlete_df[athlete_df['Medal'] == 'Bronze']

plt.figure(figsize=(12, 8)) 

sns.kdeplot(gold_medalists['Age'].dropna(), color='gold', label='Gold Medalists', linewidth=2, fill=True, alpha=0.3)
sns.kdeplot(silver_medalists['Age'].dropna(), color='silver', label='Silver Medalists', linewidth=2, fill=True, alpha=0.3)
sns.kdeplot(bronze_medalists['Age'].dropna(), color='brown', label='Bronze Medalists', linewidth=2, fill=True, alpha=0.3)

plt.title('Age Distribution of Medalists', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(title='Medal Type', fontsize=12)

plt.show()
```

```python

x = []
name = []
famous_sports = ['Basketball', 'Judo', 'Football', 'Tug-of-war', 'Athletics', 'Swimming', 'Badminton', 'Sailing', 
                 'Gymnastics', 'Art Competition', 'Handball', 'Weightlifting', 'Water Polo', 'Wrestling', 'Hockey', 
                 'Fencing', 'shooting', 'Boxing', 'Rowing', 'Taekwondo', 'Cycling', 'Diving', 'Canoeing', 'Tennis', 
                 'Golf', 'Softball', 'Archery', 'Volleyball', 'Synchronized Swimming', 'Table Tennis', 'Baseball', 
                 'Rhytmic Gymnastics', 'Rugby Sevens', 'Beach Volleyball', 'Triathlon', 'Rugby', 'Polo', 'Ice Hockey']

# Increase figure size to enhance visibility
plt.figure(figsize=(14, 10))

# Loop through each sport and plot the KDE for non-empty age data
for sport in famous_sports:
    temp_df = athlete_df[athlete_df['Sport'] == sport]
    age_data = temp_df[temp_df['Medal'] == 'Gold']['Age'].dropna()  # Filter age data
    if not age_data.empty:  # Check if the sequence is non-empty
        sns.kdeplot(age_data, label=sport, fill=True, alpha=0.2)

# Customize the plot for better readability
plt.title('Age Distribution of Gold Medalists Across Different Sports', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(title='Sport', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

men = athlete_df[athlete_df['Sex'] == 'M'].groupby('Year')['Name'].count().reset_index()
women = athlete_df[athlete_df['Sex'] == 'F'].groupby('Year')['Name'].count().reset_index()

final = men.merge(women, on='Year', suffixes=('_Male', '_Female'))

final.rename(columns={'Name_Male': 'Male', 'Name_Female': 'Female'}, inplace=True)

fig = px.line(final, x='Year', y=['Male', 'Female'], 
              title='Olympics Participation of Male and Female Athletes Over the Years',
              labels={'value': 'Number of Athletes', 'variable': 'Gender'})

fig.show(renderer="iframe_connected")
```
